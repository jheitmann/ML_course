```python
import csv
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
TRAIN_HISTORY_PATH = "results/train_history.csv"
```

```python
with open(TRAIN_HISTORY_PATH, 'r') as csvf:
    csvreader = csv.reader(csvf, delimiter=',')
    csvrows = [row for row in csvreader]

legend = np.array(csvrows[0])
data = np.array(csvrows[1:])
print(data.shape)

epochs = data[:,0]
metrics = data[:,1:]
print(epochs.shape, metrics.shape)
```

```python
plt.plot(epochs, metrics)
plt.xlabel("epoch")
plt.ylabel("metrics value")
plt.title("metrics evolution during training")
plt.show()
```
